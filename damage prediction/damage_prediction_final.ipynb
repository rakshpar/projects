```python
%matplotlib inline


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.mpl_style', 'default')
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
building_struct_df_one_hot_encoded = pd.get_dummies(data=building_struct_df,columns=['land_surface_condition','foundation_type','roof_type',
                                                                     'ground_floor_type','other_floor_type','position',
                                                                    'plan_configuration','condition_post_eq'],drop_first=True)
```

```python
building_struct_df_one_hot_encoded.columns
```

```python
building_struct_df_one_hot_encoded.head(5)
```

```python
result_inner = pd.merge(final_data, building_struct_df_one_hot_encoded, how='inner', on=['building_id'])
```

```python
result_inner.shape
```

```python
result_inner_ids_removed = result_inner.drop(['building_id','district_id_x','district_id_y','vdcmun_id_x','vdcmun_id_y'],axis=1)
```

```python
result_inner_ids_removed.astype('float').head()
```

```python
X = result_inner_ids_removed.iloc[:,1:74]
Y = result_inner_ids_removed.iloc[:,0]
test_size = 0.1
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size
                                                    , random_state=seed)
```

```python
model = XGBClassifier(n_estimators=500,max_depth=6,gamma=0.1,reg_lambda=10,learning_rate=0.1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred.astype('float')]
```

```python
print(len(y_pred))
```

```python
accuracy = accuracy_score(y_test.astype('float'), predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from sklearn import metrics
print(metrics.f1_score(y_test.astype('float'), predictions,average='weighted'))
```
